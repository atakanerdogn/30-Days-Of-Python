```python
# split data into two parts, defaults and non defaults
default = post[post['TARGET'] == 1]
non_default = post[post['TARGET'] == 0]
```

```python
default.head()
```

```python
non_default.head()
```

```python
# data analysis using plots

# gender
plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)

sns.countplot(default['CODE_GENDER'])
plt.title('Default Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.subplot(1,2,2)
sns.countplot(non_default['CODE_GENDER'])
plt.title('Non-Default Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
#education status
plt.figure(figsize=[20,10])
plt.subplot(1,2,1)

sns.countplot(default['NAME_EDUCATION_TYPE'])
plt.title('Default Count by Education Type')
plt.xlabel('Education Type')
plt.ylabel('Count')

plt.subplot(1,2,2)
sns.countplot(non_default['NAME_EDUCATION_TYPE'])
```

```python
plt.figure(figsize=[20,10])

sns.boxplot(x='TARGET', y='AMT_CREDIT', data=post)
plt.title('Credit Amount Distribution by Default Status')
plt.xlabel('Default Status (0: Non-Default, 1: Default)')
plt.ylabel('Credit Amount')

plt.tight_layout()
plt.show()
```
